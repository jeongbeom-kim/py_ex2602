```python
import matplotlib.pyplot as plt

# Figure 생성
fig = plt.figure()

# 1행 2열 subplot
a1 = fig.add_subplot(1, 2, 1)
a2 = fig.add_subplot(1, 2, 2)

# 첫 번째 subplot
x1 = range(0, 10)
y1 = range(5, 15)
a1.plot(x1, y1, 'rd')  # 빨간색 점

# 두 번째 subplot
x2 = range(0, 50)
y2 = range(0, 50)
a2.plot(x2, y2, 'y^')  # 노란색 삼각형 마커

# 그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")

# figure 생성
fig = plt.figure()

# 1행 1열 subplot
a = fig.add_subplot(111)

# 데이터
x = [0, 1]             # bar 그래프는 x좌표 index 필요
y = [3, 4]

labels = ['english', 'math']
colors = ['red', 'blue']

# 막대 그래프
a.bar(x, y, tick_label=labels, color=colors)

# 그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")

# figure 생성
fig = plt.figure()

# 1행 1열 subplot
a = fig.add_subplot(111)

# 데이터
x = [0, 1]             # bar 그래프는 x좌표 index 필요
y = [3, 4]

labels = ['english', 'math']
colors = ['red', 'blue']

# 막대 그래프
a.barh(x, y, tick_label=labels, color=colors)

# 그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")

fig = plt.figure()
a = fig.add_subplot(111)

x = [0, 1]
y = [3, 4]

labels = ['english', 'math']
colors = ['red', 'blue']

# 막대 그래프
a.bar(x, y, tick_label=labels, color=colors)

# 막대 위 값 표시
for i, v in enumerate(y):
    a.text(x[i], v + 0.1, str(v), ha='center', va='bottom')

plt.show()
```
