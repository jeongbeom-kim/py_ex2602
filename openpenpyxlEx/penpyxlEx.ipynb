```python
import openpyxl
from openpyxl import Workbook, load_workbook

print(openpyxl.__version__)

# 새 파일 생성
w1 = Workbook()
ws = w1.active
w1.save('test.xlsx')

# 파일 불러오기
w2 = load_workbook('test.xlsx')
ws2 = w2.active

ws2['A1'] = 'hello'
ws2['C3'] = 'bye'

# D5~F8 python 채우기
for row in ws2['D5':'F8']:
    for cell in row:
        cell.value = 'python'

# A1~C3 숫자 증가
num = 2
for row in ws2['A1':'C3']:
    for cell in row:
        cell.value = num
        num += 1

w2.save('test.xlsx')
```

```python
from openpyxl import Workbook

item1 = {'품목': '바나나', '가격': '30000', '수량': '1', '날짜': '2026년 02월 23일'}
item2 = {'품목': '수박', '가격': '40000', '수량': '1', '날짜': '2026년 02월 23일'}
item3 = {'품목': '오렌지', '가격': '50000', '수량': '1', '날짜': '2026년 02월 23일'}
item4 = {'품목': '딸기', '가격': '80000', '수량': '1', '날짜': '2026년 02월 23일'}

wb = Workbook()
ws = wb.active

ws['A1'] = '품목'
ws['B1'] = '가격'
ws['C1'] = '수량'
ws['D1'] = '날짜'

items = [item1, item2, item3, item4]

i = 2
for item in items:
    ws['A' + str(i)] = item['품목']
    ws['B' + str(i)] = item['가격']
    ws['C' + str(i)] = item['수량']
    ws['D' + str(i)] = item['날짜']
    i += 1

wb.save('items.xlsx')
```
